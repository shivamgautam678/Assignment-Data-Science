```python
# Q1. What is Abstraction in OOPs? Explain with an example.

# Abstraction in object-oriented programming (OOP) is a fundamental principle that focuses on representing the essential features
# and behaviors of an object while hiding unnecessary details. It allows us to create abstract classes and interfaces that define
# a blueprint for derived classes to implement.

# Abstraction helps in managing complexity by breaking down a system into smaller, more manageable parts. It allows us to create 
# high-level, conceptual models that hide the implementation details. The main goal of abstraction is to provide a simplified view
# of an object or system, emphasizing only the necessary and relevant information.

# For example, consider the concept of a "Vehicle." A vehicle can be an abstract class that defines common properties and 
# behaviors shared by all types of vehicles, such as a car, a bike, or a truck. The abstract class "Vehicle" may have methods
# like "start_engine," "stop_engine," and "accelerate" without specifying how each method is implemented. The derived classes,
# like "Car" or "Bike," will provide their specific implementations for these methods. The abstract class abstracts away the 
# implementation details and provides a generalized representation of a vehicle.


# Q2. Differentiate between Abstraction and Encapsulation. Explain with an example.

# Abstraction and encapsulation are two important concepts in object-oriented programming, but they serve different purposes.

# Abstraction focuses on hiding unnecessary details and providing a simplified representation of an object or system. 
# It allows us to define abstract classes and interfaces that specify the behavior and properties without getting into
# the specific implementation. Abstraction deals with the "what" and "why" of an object or system.

# Encapsulation, on the other hand, is about bundling related data and behavior into a single unit, known as a class.
# It allows us to control the access to the internal state of an object, protecting it from external interference.
# Encapsulation deals with the "how" of an object or system.

# Here's an example to illustrate the difference:

# Abstraction:
# Let's consider the concept of a "Shape." A shape can be abstractly defined with properties like "area" and "perimeter." 
# We can have an abstract class called "Shape" that provides these properties as abstract methods, without specifying the
# implementation details. Derived classes like "Circle" or "Rectangle" will inherit from the "Shape" class and provide their 
# specific implementations for calculating the area and perimeter.

# Encapsulation:
# Suppose we have a class called "Person" that represents a person's information. It may have private data members like name,
# age, and address. We encapsulate these data members by providing public methods like "setName," "setAge," and "setAddress" 
# to access and modify them. This way, we can control how the internal state of the "Person" object is accessed and ensure that
# it follows certain rules or restrictions.



# Q3. What is the abc module in Python? Why is it used?

# The `abc` module in Python stands for "Abstract Base Classes." It is a built-in module that provides mechanisms for defining 
# abstract base classes. Abstract base classes are classes that are meant to be inherited by other classes and serve as a 
# blueprint for their implementations.

# The `abc` module allows us to create abstract base classes by using the `ABC` class as a metaclass or by using the
# `@abstractmethod` decorator. By defining abstract methods in these classes, we can specify a common interface that 
# the derived classes must implement.

# The `abc` module is used to enforce a certain structure or behavior across multiple classes. It helps in achieving
# abstraction and provides a way to define a common interface for a group of related classes.



# Q4. How can we achieve data abstraction?

# Data abstraction in object-oriented programming refers to the concept of hiding the internal details

#  of an object and providing only the essential information to the outside world. It allows us to separate the interface
#     from the implementation.

# In many object-oriented languages, including Python, data abstraction can be achieved through the use of abstract classes
# and interfaces. Abstract classes define the abstract methods that must be implemented by derived classes, while interfaces
# specify a contract that implementing classes must adhere to.

# To achieve data abstraction:

# 1. Define an abstract class or interface that declares the desired methods and properties without providing their
# implementations.
# 2. In the abstract class, mark the methods as abstract using the `@abstractmethod` decorator or by using the `ABC` metaclass.
# 3. Create derived classes that inherit from the abstract class or implement the interface.
# 4. Implement the abstract methods in the derived classes, providing the necessary logic and functionality.

# By following this approach, you can achieve data abstraction by defining a common interface that derived classes must
# implement, while hiding the internal implementation details.




# Q5. Can we create an instance of an abstract class? Explain your answer.

# No, we cannot create an instance of an abstract class. An abstract class is designed to be a blueprint or template for other
# classes to inherit from and implement its abstract methods. It cannot be instantiated on its own because it may have incomplete
# or undefined methods.

# An abstract class can have both abstract methods (marked with `@abstractmethod`) and concrete methods (with implementations).
# However, an abstract class must have at least one abstract method for it to be considered abstract.

# To use an abstract class, you need to create a derived class that inherits from the abstract class and provides implementations'
# for all the abstract methods. The derived class can then be instantiated and used.

# Attempting to instantiate an abstract class directly will result in a `TypeError`. Abstract classes are meant to be extended 
# and provide a common interface, but they cannot be instantiated because they are incomplete and lack the necessary 
# implementations.
```
